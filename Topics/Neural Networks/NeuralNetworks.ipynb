# Neural Networks
While simpler algorithms like Linear Regression or Decision Trees should the first choice, Neural Networks excel when the number of features is very large or the problem is complex and highly non-linear.

Neural Networks attempt to bridge the gap between what is easy for computers and what is easy for humans:
- Easy for Computers / Hard for Humans: Large-Scale arithmetic and complex calculations
- Easy for Humans / Hard for Computers: Recognising a person from a side profile or understanding speech in a noisy room

The realisation that adding hidden layers between the input and output allowed to learn non-linear patterns by stacking neurons, we create a Multi-Layered Perceptron (MLP). Modern Deep Learning is essentially just an MLP with many, many hidden layers.

## Anatomy of an Artifical Neuron
A single neuron which is often called a node or unit performs a two-step calculation which takes in several inputs and produces one output.
- Inputs ($x$): These are the features from the data like pixel values, temperature, etc!
- Weights ($w$): These represent the strength of the connection where each input has its own weight.
- Bias ($b$): An extra tunable parameter that allows the neuron to shift the activation function up or down, which acts like an intercept in a linear equation.

### Step 1: The Weighted Sum
The first thing the neuron does is calculate a linear combination of the inputs and weights, then adds the bias. This is called the net input or $z$
- The Formula: $z = \sum_{i=1}^{n} (x_i \cdot w_i) + b$
- In vector notation, which is actually how computers calculate it: $z = w^T \cdot x + b$

### Step 2: The Activation Function
The value $z$ can be any number from negative infinity to positive infinity. To make it useful, we pass it through an Activation Function $\sigma$
- Purpose: It introduces Non-Linearity where without this, a neural network would just be a giant linear regression model no matter how many layers are added
- The Ouput ($a$): $a = \sigma(z)$

### Common Activation Functions
The slides introduce the most common squashing functions used in the hidden layers:
- Sigmoid: Squashes values between 0 and 1 $\sigma(z) = \frac{1}{1 + e^{-z}}$
- ReLU (Rectified Linear Unit): The modern standard where it outputs the value if its positive and 0 if its negative $f(z) = \max(0, z)$

## Architecture of the Multi-Layered Perceptron (MLP)
A standard Neural Network is organised into three types of layers. Information typically flows into one direction, called Forward Propagation.
- Input Layer: Receives the raw data. There is one node for every feature in the dataset.
- Hidden Layers: These are the "black box" layers between input and output. They perform intermediate calculations to extract complex patterns. A network can have one or many (Deep Learning)
- Ouput Layer: Produces the final prediction (0 or 1 for classification or a continuous number for regression)

### Fully Connected Dense Layers
In an MLP, layers are usually fully connected. this means:
- Every neuron in layer $i$ is connected to every neuron in layer $i + 1$
- Each connection has its own unique weight $w_{ij}$
- The total number of parameters (weights + biases) grows very quickly as you add neurons

### Why Hidden Layers?
The term hidden simply means these layers do not interact with the outside world directly
- Mathematical Power: A single hidden layer with enough neurons can approximate almost any continuous function
- Feature Hierarchies: Early hidden layers might detect simple edges, while deeper layers combine those edges to detect shapes or objects

### Forward Propagation Maths
To calculate the output of a whole layer, we use matrix multiplication.
- The Vectorised Formula: $a^{(l)} = \sigma(W^{(l)} \cdot a^{(l-1)} + b^{(l)})$
- Matrix of weights for layer $l$: $W^{(l)}$
- Vector of biases for layer $l$: $b^{(l)}$
- Activation function applied element-wise: $\sigma$

## Neural Network Layer Types

| Layer Type | Purpose | Activation Function |
|------------|---------|---------------------|
| Input | Passes data into the network. | None (usually) |
| Hidden | Learns non-linear features. | Usually ReLU |
| Output | Provides the final result. | Sigmoid (Binary) or Softmax (Multi-class) |

# Loss Functions and Optimisation
This is the feedback loop. If Forward Propagation is the network making a guess, this section is about the network being told how wrong it is and adjusting its weights to do better next time.

## The Loss Function
The Loss Function measures the distance between the networks prediction and the actual target
- Goal: We want to minimise this value. A loss of 0 means a perfect prediction
- Mean Squared Error (MSE): Used for Regression $MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
- Binary Cross-Entropy: Used for Binary Classification $L = -\frac{1}{n} \sum [y \log(\hat{y}) + (1-y) \log(1-\hat{y})]$

## Gradient Descent
How do we actually lower the loss? We use Gradient Descent
- Imagine you are on a foggy mountain (the Loss Curve) and want to find the valley (minimum loss). You feel the slope under your feet and take a step in the steepest downward direction
- The Gradient: The derivative of the loss with respect to the weights. It tells us which way is up $\nabla L = \frac{\partial L}{\partial w}$

## The Learning Rate
This is the Hyperparameter represented by the symbol eta $\eta$. It determines how big of a step the network takes.
- Too Small: Training takes forever and might get stuck in a shallow dip
- Too Large: The network overshoots the valley and might never converge

## Backpropagation
This is the magic of Neural Networks. Since the error happens at the Output, we need a way to tell the neurons in the Hidden Layers how much they contributed to that mistake.
- Using the Chain Rule from calculus, the error is sent backward through the network.
- Every weight in every layer is updated slightly $w_{new} = w_{old} - \eta \cdot \frac{\partial L}{\partial w}$

## Gradient Descent Terminology

| Term | Analogy | Mathematical Role |
|------|---------|-------------------|
| Loss Function | The Scoreboard | Measures error: $L(y, \hat{y})$ |
| Gradient | The Slope | Direction to move: $\frac{\partial L}{\partial w}$ |
| Learning Rate | Step Size | Speed of update: $\eta$ |
| Backprop | Apportioning Blame | Distributing error to previous layers |

## Epochs and Batches
You dont just show the network the data once. You show it repeatedly.
- Epoch: One full pass of the entire dataset through the neural network
- Batch Size: The number of training examples processed before the internal paramters like weights are updated.
    - Stochastic: Batch Size = 1 (Very noisy but fast)
    - Mini-Batch: Batch Sizr = 32 or 64 (The sweet spot for modern hardware)
    - Batch: Process the entire dataset at once (Very stable but memory-intensive)

### Overfitting
Overfitting is the biggest challenge in Neural Networks as its when a network memorises the training data instead of the underlying patterns, a sign of such would be low error on Training Data but High Error on Validation/Test Data

### Underfitting
The opposite of overfitting. The model is too simple as it does not have enough layers or neurons to even learn the training data. This can be seen if there is a high error on both Training and Validation data

### Preventing Overfitting
- Early Stopping: Stop training the moment validation error starts to rise
- Dropout: Randomly turn off neurons during training so the network doesnt rely too heavily on any single node
- Data Augmentation: Artifically increasing the size of the dataset like flipping or roatating images

## Training Terminology

| Term | Definition | Goal |
|------|------------|------|
| Epoch | One full pass of the data. | Iterative improvement. |
| Batch | Subset of data per update. | Computational efficiency. |
| Overfitting | Memorizing noise/details. | Avoid! (Worsens generalization). |
| Underfitting | Failing to find a pattern. | Avoid! (Model is too weak). |

## Logits and Softmax
Before the very last step, the output neurons produce raw numbers called Logits
- Definition: The unconstrained real numbers generated by the last layer before an activation function is applied
- Range: $(-\infty, +\infty)$
- Example: For a 3-class problem (Cat, Dog, Bird), the logits might be $[2.0, 1.0, 0.1]$. These arent probabilities yet because they dont sum to 1.

## The Softmax Function
To turn those raw logits into probabilities, we use Softmax. It squashes the logits so that they are all between 0 and 1 and their total sum is exactly 1.
- The Formula: $\sigma(z)_i = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}$
- Result: Our Example $[2.0, 1.0, 0.1]$ becomes $[0.66, 0.24, 0.10]$. The network is 66% sure the image is a Cat.

## Numerical Stability & TensorFlow
A key practical tip from the slides
- The Problem: Calculating $log$ on very small probabilities causes rounding errors
- The Solution: In libraries like TensorFlow, it is better to calculate the loss directly from the Logits rather than the probabilities.
- Parameter: You set `from_logits=true` in the loss function. This tells the computer to handle the maths internally for better precision.
