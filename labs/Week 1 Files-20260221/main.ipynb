```python
# Here we can define the data points used for plotting
l2x = np.array([1, 5, 6, 10])
l2y = np.array([0, 100, 175, 250])

l1x = np.array([0, 6])
l1y = np.array([0, 250])

# The 'o' means rings, so it can remove the plotted lines
plt.subplot(2,1,1)
plt.plot(l1x, l1y, l2x, l2y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("House Data Charts", loc='left')
plt.grid(axis='y')

plt.subplot(2,1,2)
plt.scatter(l1x, l1y)
plt.show()
```

## Calculations


```python
student_data["average_score"] = (student_data["math_score"] + student_data["science_score"]) / 2

print(student_data[["name", "average_score"]])
print("High Performers: ", student_data[student_data["average_score"] > 80])
```
